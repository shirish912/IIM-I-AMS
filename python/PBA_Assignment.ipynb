```python
import folium
from folium.plugins import HeatMap
```

```python
# Create pandas DataFrame from your data
data = {
    'CustomerID': [1001, 1002, 1003, 1004],
    'Latitude': [37.7749, 34.0522, 41.8819, 30.2672],
    'Longitude': [-122.4194, -118.2437, -87.6231, -97.7401],
    'Sales': [1000, 800, 1500, 1200]
}

df = pd.DataFrame(data)

# Get center coordinates (average of all latitudes and longitudes)
center_latitude = df['Latitude'].mean()
center_longitude = df['Longitude'].mean()

# Create a base map
map = folium.Map(location=[center_latitude, center_longitude], zoom_start=7)

# Add markers with customer IDs as popups
for i, row in df.iterrows():
  marker = folium.Marker([row['Latitude'], row['Longitude']], popup=f"Customer ID: {row['CustomerID']}")
  marker.add_to(map)

# Add heatmap layer (optional)
if 'Sales' in df.columns:
  # Adjust radius and max_zoom for better heatmap visualization based on data
  heatmap = folium.HeatMap(data=list(zip(df['Latitude'], df['Longitude'], df['Sales'])), radius=15, max_zoom=12)
  heatmap.add_to(map)

# Display the map
map
```
