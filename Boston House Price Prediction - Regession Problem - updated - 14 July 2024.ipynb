## Linear Regression using sklearn

```python
from sklearn.linear_model import LinearRegression

# LinearRegression is a CLASS, hence it has to be instantiated before use
LR = LinearRegression()
```

```python
LR.fit(X_train, y_train)  # model training step
# "fitting" always means "learning something from data"
# So, in this case, the LR is learning the "optimal values of the model coefficinats"
# which are the slope(s) and intercept
```

```python
LR.predict(X_test)   # model's predictions on the test set
```

```python
LR.score(X_test, y_test)   # R^2 for the LR model on the test set
```

```python
LR.score(X_train, y_train)   # R^2 for the LR model on the test set
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
coeff_df = pd.DataFrame(LR.coef_, index=X.columns, columns=['Model Coeff'])
coeff_df
```

## Using the OLS Normal Equation

```python
ones = np.ones(X.shape[0]).reshape(-1, 1)
X_new = np.hstack([ones, X])
X_new
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)
```
