# Gradient Descent Algorithm

### Overview
Gradient Descent is an optimization algorithm used to minimize the loss function in machine learning and deep learning models. It iteratively adjusts the model parameters to find the values that minimize the loss function.

### Mathematical Formulation

1. **Loss Function (Objective Function)**: 
   The loss function $ L(w) $ measures the performance of the model as a function of the model coefficients $ w $. For a simple linear regression, the loss function is often the Mean Squared Error (MSE):

   $$
   L(w) = \frac{1}{2N} \sum_{i=1}^N (\hat{y}_i - y_i)^2
   $$

   where:
   - $ N $ is the number of training examples.
   - $ \hat{y}_i $ is the predicted value, $ \hat{y}_i = w_0 + w_1 x_i $ for linear regression.
   - $ y_i $ is the actual output.

   **Reason for $ \frac{1}{2N} $ instead of $ \frac{1}{N} $**:
   The factor $ \frac{1}{2} $ is included for mathematical convenience. It simplifies the derivative of the loss function when performing gradient descent, as the 2 in the exponent of the squared term cancels out when differentiated.

2. **Gradient Descent Update Rule**:
   The goal is to update the coefficients $ w $ to minimize $ L(w) $. This is done by taking steps proportional to the negative of the gradient of the loss function. The update rule is:

   $$
   w_j = w_j - \alpha \frac{\partial L(w)}{\partial w_j}
   $$

   where:
   - $ \alpha $ is the learning rate.
   - $ \frac{\partial L(w)}{\partial w_j} $ is the partial derivative of the loss function with respect to $ w_j $.

3. **Gradient Calculation**:
   For each coefficient $ w_j $, the partial derivative of the loss function is computed. For linear regression, it is:

   $$
   \frac{\partial L(w)}{\partial w_j} = \frac{1}{N} \sum_{i=1}^N \left( \hat{y}_i - y_i \right) x_{ij}
   $$

   Here, $ x_{ij} $ is the $ j $-th feature of the $ i $-th training example.

### Steps in Gradient Descent

1. **Initialize Coefficients**: Start with initial guesses for $ w $.
2. **Compute Gradient**: Calculate the gradient of the loss function with respect to each coefficient.
3. **Update Coefficients**: Adjust the coefficients using the update rule.
4. **Repeat**: Continue until convergence, i.e., when the change in the loss function is below a threshold.

### Types of Gradient Descent

1. **Batch Gradient Descent**: Uses the entire dataset to compute the gradient. This is computationally expensive for large datasets.
2. **Stochastic Gradient Descent (SGD)**: Uses one training example at a time to compute the gradient. This is faster but introduces more noise.
3. **Mini-Batch Gradient Descent**: Uses a small batch of training examples to compute the gradient. It strikes a balance between the stability of batch gradient descent and the speed of SGD.

### Applications in ML and DL

1. **Linear Regression**: Used to minimize the loss function and find the best fit line.
2. **Logistic Regression**: Used to find the coefficients that best separate the classes in binary classification problems.
3. **Neural Networks**: Used to optimize weights and biases in deep learning models. Backpropagation, a special case of gradient descent, is used to compute gradients for neural networks.
4. **Support Vector Machines**: Gradient descent can be used to find the optimal hyperplane that maximizes the margin between classes.

### Example

Consider a simple linear regression model with one feature:

$$
\hat{y}_i = w_0 + w_1 x_i
$$

The loss function is:

$$
L(w) = \frac{1}{2N} \sum_{i=1}^N (w_0 + w_1 x_i - y_i)^2
$$

The gradients are:

$$
\frac{\partial L(w)}{\partial w_0} = \frac{1}{N} \sum_{i=1}^N (w_0 + w_1 x_i - y_i)
$$

$$
\frac{\partial L(w)}{\partial w_1} = \frac{1}{N} \sum_{i=1}^N (w_0 + w_1 x_i - y_i) x_i
$$

The update rules are:

$$
w_0 = w_0 - \alpha \frac{1}{N} \sum_{i=1}^N (w_0 + w_1 x_i - y_i)
$$

$$
w_1 = w_1 - \alpha \frac{1}{N} \sum_{i=1}^N (w_0 + w_1 x_i - y_i) x_i
$$

By iteratively applying these updates, the coefficients $ w_0 $ and $ w_1 $ are adjusted to minimize the loss function, thus fitting the best possible line to the data.

### Summary

Gradient Descent is a fundamental algorithm in machine learning and deep learning for optimizing models. It is widely used due to its simplicity and effectiveness in finding the minimum of loss functions, which is crucial for training accurate predictive models.

# Different Types of Error Functions in Regression

In regression analysis, various error functions are used to evaluate the performance of a model. Each error function has its own characteristics and is suitable for different scenarios. Here are the common error functions:

### 1. Sum of Errors (SE)

$$
SE = \sum_{i=1}^N (y_i - \hat{y}_i)
$$

- **Description**: The sum of the differences between the actual and predicted values.
- **Use Case**: Rarely used as an evaluation metric because positive and negative errors can cancel each other out.

### 2. Sum of Absolute Errors (SAE)

$$
SAE = \sum_{i=1}^N |y_i - \hat{y}_i|
$$

- **Description**: The sum of the absolute differences between the actual and predicted values.
- **Use Case**: Used when the cost of errors is linear, and we want to penalize all errors equally without canceling out positive and negative errors.

### 3. Sum of Squared Errors (SSE)

$$
SSE = \sum_{i=1}^N (y_i - \hat{y}_i)^2
$$

- **Description**: The sum of the squared differences between the actual and predicted values.
- **Use Case**: Commonly used in regression analysis. It heavily penalizes large errors, making it useful when large errors are particularly undesirable.

### 4. Mean Squared Error (MSE)

$$
MSE = \frac{1}{N} \sum_{i=1}^N (y_i - \hat{y}_i)^2
$$

- **Description**: The average of the squared differences between the actual and predicted values.
- **Use Case**: Widely used due to its mathematical properties, making it suitable for gradient descent optimization.

### 5. Mean Absolute Error (MAE)

$$
MAE = \frac{1}{N} \sum_{i=1}^N |y_i - \hat{y}_i|
$$

- **Description**: The average of the absolute differences between the actual and predicted values.
- **Use Case**: Preferred when outliers are not a major concern, and we want a simple, interpretable error metric.

### 6. Root Mean Squared Error (RMSE)

$$
RMSE = \sqrt{\frac{1}{N} \sum_{i=1}^N (y_i - \hat{y}_i)^2}
$$

- **Description**: The square root of the MSE. It brings the error metric back to the same unit as the original data.
- **Use Case**: Useful for comparing errors in different units or scales, often used in forecasting and time series analysis.

### 7. Sum of Logarithmic Errors (SLE)

$$
SLE = \sum_{i=1}^N \ln|y_i - \hat{y}_i|
$$

- **Description**: The sum of the natural logarithms of the absolute differences between the actual and predicted values.
- **Use Case**: Less common, used in specific applications where the logarithm of the error is more meaningful.

### 8. Root Mean Squared Logarithmic Error (RMSLE)

$$
RMSLE = \sqrt{\frac{1}{N} \sum_{i=1}^N (\ln(1 + y_i) - \ln(1 + \hat{y}_i))^2}
$$

- **Description**: The square root of the average squared logarithmic differences between the actual and predicted values.
- **Use Case**: Preferred when dealing with targets that have a wide range of values or when the focus is on relative rather than absolute differences.

### 9. Huber Loss

$$
L_\delta(y, \hat{y}) = 
\begin{cases} 
\frac{1}{2} (y - \hat{y})^2 & \text{for } |y - \hat{y}| \leq \delta \\
\delta |y - \hat{y}| - \frac{1}{2} \delta^2 & \text{for } |y - \hat{y}| > \delta 
\end{cases}
$$

- **Description**: Combines the best properties of MSE and MAE. It is quadratic for small errors and linear for large errors.
- **Use Case**: Useful when dealing with data containing outliers. The parameter $ \delta $ determines the threshold between quadratic and linear behavior.

## Preferred Use Cases for Each Error Function

1. **MSE**: Preferred for general regression problems, especially when large errors need to be heavily penalized. Suitable for gradient-based optimization methods.

2. **MAE**: Preferred when the model needs to be robust to outliers or when errors need to be treated equally. It provides a more interpretable measure of average error.

3. **RMSE**: Used in situations where it is important to have the error metric in the same units as the original data. Common in forecasting and time series analysis.

4. **RMSLE**: Suitable for problems where the target variable spans several orders of magnitude or when relative error is more important than absolute error.

5. **Huber Loss**: Ideal for regression problems with outliers. It offers a balance between the sensitivity of MSE to outliers and the robustness of MAE.

By understanding these error functions and their use cases, you can select the appropriate metric for evaluating and optimizing your regression models in different real-world scenarios.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + 0.5*np.random.randn(100, 1)
```

```python
# Function to compute the Mean Squared Error (MSE) loss
def compute_loss(X, y, w0, w1):
    N = len(y)
    y_hat = w0 + w1 * X
    L = (1 / (2 * N)) * np.sum(np.square(y_hat - y))
    return L

# Gradient descent function
def gradient_descent(X, y, w0, w1, learning_rate, iterations):
    N = len(y)
    loss_history = np.zeros(iterations)
    w0_history = np.zeros(iterations)
    w1_history = np.zeros(iterations)
    
    for i in range(iterations):
        y_hat = w0 + w1 * X
        d_w0 = (1 / N) * np.sum(y_hat - y)
        d_w1 = (1 / N) * np.sum((y_hat - y) * X)
        
        w0 = w0 - learning_rate * d_w0
        w1 = w1 - learning_rate * d_w1
        
        loss_history[i] = compute_loss(X, y, w0, w1)
        w0_history[i] = w0
        w1_history[i] = w1
        
        if i % 100 == 0:
            print(f"Iteration {i}: Loss {loss_history[i]}, w0 {w0}, w1 {w1}")
    
    return w0, w1, loss_history, w0_history, w1_history
```

```python
# Initial parameters
w0 = 0
w1 = 0
learning_rate = 0.05
iterations = 1000

# Running gradient descent
w0, w1, loss_history, w0_history, w1_history = gradient_descent(X, y, w0, w1, learning_rate, iterations)
```

```python
# Plotting the results
```

```python
# Plotting the original data with the best-fit line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', marker='o', label='Data points')
plt.plot(X, w0 + w1 * X, color='red', label=f'Best fit line: y = {w0:.2f} + {w1:.2f}X')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Original Data with Best Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot of Loss vs. Iterations
plt.figure(figsize=(8,5))
plt.plot(range(iterations), loss_history, 'b.')
plt.xlabel('Iterations')
plt.ylabel('Loss (MSE)')
plt.title('Loss Reduction Over Iterations')
plt.grid(True)
plt.show()
```

```python
# Plot of w0 vs. Iterations and w1 vs. Iterations side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Loss vs. w0
ax1.plot(w0_history, loss_history, 'm.')
ax1.set_xlabel('w0')
ax1.set_ylabel('Loss (MSE)')
ax1.set_title('Loss vs. w0')
ax1.grid(True)

# Loss vs. w1
ax2.plot(w1_history, loss_history, 'c.')
ax2.set_xlabel('w1')
ax2.set_ylabel('Loss (MSE)')
ax2.set_title('Loss vs. w1')
ax2.grid(True)

plt.show()
```

```python
# Plot of w0 vs. Iterations and w1 vs. Iterations side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# w0 vs. Iterations
ax1.plot(range(iterations), w0_history, 'r.')
ax1.set_xlabel('Iterations')
ax1.set_ylabel('w0')
ax1.set_title('w0 Over Iterations')
ax1.grid(True)

# w1 vs. Iterations
ax2.plot(range(iterations), w1_history, 'g.')
ax2.set_xlabel('Iterations')
ax2.set_ylabel('w1')
ax2.set_title('w1 Over Iterations')
ax2.grid(True)

plt.show()
```

```python

# 3D plot of Loss vs. w0 and w1
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(w0_history, w1_history, loss_history, 'b.')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Loss (MSE)')
ax.set_title('Loss Surface')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Function to compute the Mean Squared Error (MSE) loss
def compute_loss(X, y, w0, w1):
    N = len(y)
    y_hat = w0 + w1 * X
    L = (1 / (2 * N)) * np.sum(np.square(y_hat - y))
    return L

# Creating a meshgrid for w0 and w1
w0_values = np.linspace(-5, 10, 100)
w1_values = np.linspace(-15, 10, 100)
w0_mesh, w1_mesh = np.meshgrid(w0_values, w1_values)
L_mesh = np.zeros_like(w0_mesh)

# Compute the loss for each combination of w0 and w1
for i in range(w0_mesh.shape[0]):
    for j in range(w0_mesh.shape[1]):
        L_mesh[i, j] = compute_loss(X, y, w0_mesh[i, j], w1_mesh[i, j])

# 3D plot of L vs. w0 and w1
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(w0_mesh, w1_mesh, L_mesh, cmap='viridis')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Loss (MSE)')
ax.set_title('Loss Surface')
plt.show()
```
